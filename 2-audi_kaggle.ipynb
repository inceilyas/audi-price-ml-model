```python
model=LinearRegression()
```

```python
model
```

```python
y=df["Price(£)"]
x=df.drop(["Price(£)"],axis=1)
```

```python
model.fit(x,y)
```

```python
model.predict([[2018,44000,1.6,114.398422,1,2499.166667,1,0,1,1,0]])
```

```python
model.score(x,y)
```

```python
y_pred=model.predict(x)
```

```python
import numpy as np

def meanScore(y, y_pred):
    return np.mean(np.abs((y - y_pred) / y)) * 100

```

```python
meanScore(y,y_pred)
```
